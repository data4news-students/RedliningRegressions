```python
import pandas as pd
dfVeg = pd.read_csv("CD_VegCover.csv")
dfTemp = pd.read_csv("CD_Temperature.csv")
dfmerged = pd.merge(dfVeg, dfTemp, on='Geography', how='left')
```

```python
dfIndex = pd.read_csv("CD_HeatIndex.csv")
dfmerged2 = pd.merge(dfmerged, dfIndex, on='Geography', how='left')
dfmerged2


dfOzone= pd.read_csv("CD_OzoneCorrect.csv")
dfmerged3 = pd.merge(dfmerged2, dfOzone, on='Geography', how='left')
dfmerged3
```

```python
%%R -i dfmerged3

ggplot(dfmerged3) +
    aes(x=Mean_ppb, y=Score) +
    geom_point(size=5, alpha=.5) +
    theme_minimal() +
    geom_smooth(method='lm')
```
