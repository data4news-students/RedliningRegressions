```python
import pandas as pd
df = pd.read_csv("sample.csv") #this has some merging from my scratchpad
dfHosp = pd.read_csv("CD_Hospitalizations.csv")

df2 = pd.merge(df, dfHosp, on='Geography', how='left')
df2
```

```python
dfPoverty = pd.read_csv("CD_Poverty.csv")
dfCombined = pd.merge(df2, dfPoverty, on='Geography', how='left')
```

```python
%%R
# load the dataset
data(dfCombined)

# run a simple linear regression of mpg (dependent variable) on wt (independent variable)
fit <- lm("HospRate ~ Percent_Poverty + Percent + Score + Degrees" , data = dfCombined)

# summarize the regression results
summary(fit)


```

```python
%%R
# load the dataset
data(dfCombined)


# run a simple linear regression of mpg (dependent variable) on wt (independent variable)
fit <- lm("HospRate ~ Degrees" , data = dfCombined)

# summarize the regression results
summary(fit)
```

<B>Shows me that its very clearly not just surface temperature that impacts Heat related hospitalizations

```python
%%R
# load the dataset
data(dfCombined)


# run a simple linear regression of mpg (dependent variable) on wt (independent variable)
fit <- lm("HospRate ~ Percent + Score + Degrees" , data = dfCombined)

# summarize the regression results
summary(fit)
```

```python
%%R
# load the dataset
data(dfCombined)


# run a simple linear regression of mpg (dependent variable) on wt (independent variable)
fit <- lm("HospRate ~ I(Percent**2) + I(Score**2) + I(Degrees**2) + I(Percent_Poverty**2)" , data = dfCombined)

# summarize the regression results
summary(fit)
```

```python
%%R
# load the dataset
data(dfCombined)


# run a simple linear regression of mpg (dependent variable) on wt (independent variable)
fit <- lm("Score ~ I(Percent**2) + I(Degrees**2) + I(Percent_Poverty**2)" , data = dfCombined)

# summarize the regression results
summary(fit)
```

<B>wait woah that's much higher than before -- seeing what factors make for high heat vulnerability index

```python
%%R
# load the dataset
data(dfCombined)


# run a simple linear regression of mpg (dependent variable) on wt (independent variable)
fit <- lm("HospRate ~ I(log(Percent)) + I(log(Score)) + I(log(Degrees)) + I(log(Percent_Poverty))" , data = dfCombined)

# summarize the regression results
summary(fit)
```
